````python
key_qualifications_prompt_template = """
[System Instruction]
You are an expert CV and LinkedIn profile skill generator. Your goal is to analyze job descriptions and user-provided talents to create a set of highly relevant and concise skills. Focus on extracting key skills from the job descriptions and tailoring them to create a compelling professional profile.

[Instructions for Skill Generation]
1. **Analyze Job Descriptions:** Carefully read the "main job description" and the "similar job descriptions" provided below. Pay close attention to sections like "Required Qualifications," "Responsibilities," "Ideal Candidate," and "Skills." Prioritize skills mentioned frequently and those listed as essential requirements.

2. **Identify Key Skills:** Extract the core skills and competencies sought in these job descriptions. Think about the underlying abilities and expertise employers are looking for.

3. **Synthesize and Condense:** Combine related skills and rephrase them to be concise and impactful. Aim for action-oriented phrases that highlight your capabilities.  Each skill should be **no longer than 27 characters**.

4. **Integrate Unique Talents:** Review the "Unique Talents" listed below. Select **up to two** talents that are most relevant to the identified key skills and the job descriptions.  Incorporate these talents by either:
    *   **Direct Integration:** If a talent directly translates into a professional skill, use it as part of your skill list.
    *   **Skill Enhancement:**  Use your talents to refine or add nuance to the skills you generate from the job descriptions. For example, if a job description emphasizes "communication" and your talent is "Public Speaking," you might generate the skill "Effective Public Communication."

5. **Generate Six Skills:**  Create a list of **six skills** that are:
    *   Highly relevant to the job descriptions.
    *   Concise (under 27 characters).
    *   Action-oriented and impactful.
    *   Reflective of your potential and, where possible, your unique talents.

[Job Descriptions and Talents]

**Main Job Description:**
{{main_job_description_raw}}

**Similar Job Description 1:**
{{similar_job_description_raw_1}}

**Similar Job Description 2:**
{{similar_job_description_raw_2}}

**Similar Job Description 3:**
{{similar_job_description_raw_3}}

**Similar Job Description 4:**
{{similar_job_description_raw_4}}

**Unique Talents:**
{{my_talents}}


[Output Format]
Generate the six skills as a bulleted list.

"""
````

````python
def extract_big_6_skills(
    prompt_template,
    main_job_description_raw,
    similar_job_descriptions,
    my_talents
):
    try:
        # Fill all 5 JD blocks (pad if fewer)
        jd_list = similar_job_descriptions + [""] * (4 - len(similar_job_descriptions))

        # Format the prompt with all necessary fields
        prompt = prompt_template.format(
            main_job_description_raw=main_job_description_raw,
            similar_job_description_raw_1=jd_list[0],
            similar_job_description_raw_2=jd_list[1],
            similar_job_description_raw_3=jd_list[2],
            similar_job_description_raw_4=jd_list[3],
            my_talents=my_talents
        )

        # Send to LLM
        response = client.chat.completions.create(
            model="deepseek-r1-distill-llama-70b",
            messages=[{"role": "user", "content": prompt}],
            temperature=0.3,
            max_tokens=1024
        )

        return response.choices[0].message.content.strip()

    except Exception as e:
        print(f"Error generating Big 6 skills: {e}")
        return None
#---------------------------------------------------------execute
similar_job_descriptions = [
    similar_job_description_raw_1,
    similar_job_description_raw_2,
    similar_job_description_raw_3,
    similar_job_description_raw_4
]
big_6 = extract_big_6_skills(
    key_qualifications_prompt_template,
    main_job_description_raw,
    similar_job_descriptions,
    my_talents
)

print("Extracted Big 6:")
print(big_6)
````

````python
def clean_big_6_with_llm(raw_response, clean_prompt_template):
    try:
        prompt = clean_prompt_template.format(raw_response=raw_response)

        response = client.chat.completions.create(
            model="llama-3.3-70b-versatile",
            messages=[{"role": "user", "content": prompt}],
            temperature=0.1,
            max_tokens=1000
        )
        return response.choices[0].message.content.strip()

    except Exception as e:
        print(f"AI Cleaning failed: {e}")
        return []

big_6_clean = clean_big_6_with_llm(big_6, clean_big_6_prompt)
# print("Cleaned Big 6:")
print(big_6_clean)
````

````python
import re
def parse_raw_big_6_list(raw_string):
    if not raw_string:
        return []

    skills_list = []
    lines = raw_string.strip().splitlines()

    for line in lines:
        cleaned_line = re.sub(r"^[*\-]\s*", "", line).strip()
        if cleaned_line:
            skills_list.append(cleaned_line)

    return skills_list

big_6_clean_list = parse_raw_big_6_list(big_6_clean)
print(big_6_clean_list)
````

````python
def extract_job_details_json(prompt_template, main_job_description_raw):
    import re
    import json

    try:
        # Inject the job description
        filled_prompt = prompt_template.replace("{{main_job_description_raw}}", main_job_description_raw)

        # Send prompt to LLM
        response = client.chat.completions.create(
            model="deepseek-r1-distill-llama-70b",
            messages=[{"role": "user", "content": filled_prompt}],
            temperature=0.2,
            max_tokens=1500
        )

        raw_output = response.choices[0].message.content.strip()

        json_candidate = re.search(r"\{.*\}", raw_output, re.DOTALL)
        if not json_candidate:
            print("No JSON object found in the output.")
            return None

        json_text = json_candidate.group()

        parsed = json.loads(json_text)
        return parsed

    except Exception as e:
        print(f"Error extracting job details: {e}")
        return None
````

````python
# === 1. Format a role dictionary ===
def format_single_role_for_prompt(role):
    info_block = f"""`<info>`
• Job Title: {role['job_title']}
• Organization: {role['organization']}
• Industry: {role['industry']}
• Team Size: {role['team_size']}
• Description: {role['description']}
`</info>`"""

    acc_blocks = "\n".join([f"`<accomplishment>` {acc} `</accomplishment>`" for acc in role["accomplishments"]])
    return f"{info_block}\n\n{acc_blocks}"


# === 2. Generate tailored resume bullets (raw) ===
def create_tailored_experience_for_role(role, big_6):
    try:
        if not resume_role_prompt_prompt_template:
            print("Missing prompt template.")
            return None

        formatted_role_block = format_single_role_for_prompt(role)
        skills_block = ", ".join(big_6)  # safer injection for prompt
        prompt = resume_role_prompt_prompt_template \
            .replace("{{Target Skills}}", skills_block) \
            .replace("{{structured_output}}", formatted_role_block)

        response = client.chat.completions.create(
            model="llama-3.3-70b-versatile",
            messages=[{"role": "user", "content": prompt}],
            temperature=0.5,
            max_tokens=2048
        )
        return response.choices[0].message.content.strip()

    except Exception as e:
        print(f"Error generating tailored experience for role '{role['job_title']}': {e}")
        return None


# === 3. Clean up the output into final markdown ===
def clean_tailored_experience_output(raw_output, role):
    try:
        prompt = f"""
Your task is to clean and structure the following raw AI output related to a resume role. The role metadata is:

Job Title: {role['job_title']}
Organization: {role['organization']}

Here is the raw output:
\"\"\"
{raw_output}
\"\"\"

Clean and transform it as follows:
1. Remove all section headers like "## Suggested Resume Bullet Point..."
2. Keep only bullet points with alignment scores of 4 or higher
3. if there s no bullet point with alignment scores of 4 or higher in a particular role, also include bullet points with alignment scores of 3 in this role
4. Sort the bullet points in descending order of alignment score
5. Output in this format:

{{
    "title": "Job Title",
    "org": "Organization",
    "location": "City, Country",
    "dates": "Start – End",
    "bullets": [
        "Concise, high-impact bullet 1.",
        "Bullet 2.",
        "Bullet 3."
    ]
}}

Important:
- DO NOT include commentary, explanations, or markdown formatting.
- Only return the dictionary.

Use concise Markdown formatting. Do not add summaries, tags, or titles.
"""

        response = client.chat.completions.create(
            model="llama-3.3-70b-versatile",
            messages=[{"role": "user", "content": prompt}],
            temperature=0.5,
            max_tokens=2048
        )

        return response.choices[0].message.content.strip()

    except Exception as e:
        print(f"Error cleaning tailored experience output for role '{role['job_title']}': {e}")
        return None


# === 4. Run the pipeline for a single role ===
def run_tailored_experience_pipeline(role, big_6):
    print(f"Processing: {role['job_title']} @ {role['organization']}")

    try:
        raw = create_tailored_experience_for_role(role, big_6)
        if not raw:
            print(f"LLM generation failed for role: {role['job_title']} @ {role['organization']}")
            return None

        cleaned = clean_tailored_experience_output(raw, role)
        if cleaned:
            print("Final Tailored Output:\n")
            print(cleaned)
            return cleaned
        else:
            print(f"Cleaning step failed for role: {role['job_title']} @ {role['organization']}")
            return None

    except Exception as e:
        print(f"Error processing role '{role['job_title']} @ {role['organization']}': {e}")
        return None


# === 5. Batch processing all roles ===
def run_batch_experience_pipeline(roles, big_6):
    """
    Processes a batch of roles and ensures the output is correctly structured.

    Args:
        roles (list): A list of roles to process.
        big_6 (list): A list of key skills.

    Returns:
        dict: A dictionary with role keys and structured outputs.
    """
    all_outputs = {}

    for idx, role in enumerate(roles, 1):
        print(f"\n[{idx}/{len(roles)}] Running pipeline for: {role['job_title']} @ {role['organization']}")
        output = run_tailored_experience_pipeline(role, big_6)

        if output:
            try:
                # Ensure the output is a valid dictionary
                structured_output = eval(output) if isinstance(output, str) else output
                if isinstance(structured_output, dict):
                    all_outputs[f"{role['job_title']} @ {role['organization']}"] = structured_output
                else:
                    print(f"Invalid structured output for role: {role['job_title']} @ {role['organization']}")
            except Exception as e:
                print(f"Error parsing output for role: {role['job_title']} @ {role['organization']}: {e}")
        else:
            print(f"Skipping role due to errors: {role['job_title']} @ {role['organization']}")

    print("\nAll roles processed.")
    return all_outputs

from IPython.display import Markdown

def format_professional_experience_markdown(final_outputs):
    """
    Formats the professional experience section into markdown with proper formatting.

    Args:
        final_outputs (dict): The output dictionary from run_batch_experience_pipeline.

    Returns:
        str: A markdown-formatted string representing the professional experience.
    """
    output_lines = []

    for key, value in final_outputs.items():
        if not isinstance(value, dict):
            print(f"Skipping invalid entry for key '{key}': Expected dict, got {type(value).__name__}")
            continue

        title = value.get("title", "Unknown Title")
        organization = value.get("org", "Unknown Organization")
        location = value.get("location", "Unknown Location")
        dates = value.get("dates", "Unknown Dates")
        bullets = value.get("bullets", [])

        # Add header for the role
        output_lines.append(f"\n### **{title}** | *{organization}* ({dates})\n")

        # Add bullet points
        for bullet in bullets:
            output_lines.append(f"- {bullet}")

        output_lines.append("\n---\n")  # Separator between roles

    # Join the lines into one final string
    return "\n".join(output_lines)

# Example usage
final_outputs = run_batch_experience_pipeline(all_roles, big_6_clean_list)
formatted_professional_roles_output = format_professional_experience_markdown(final_outputs)

# Display the formatted output as Markdown
display(Markdown(formatted_professional_roles_output))
````

````python
def extract_project_bullets_from_combined_header(raw_output):
    import re

    lines = raw_output.strip().split("\n")
    bullets = []
    project_name = ""
    tools_line = ""

    for line in lines:
        line = line.strip()

        # Combined header: "# Project Name | **Tools**: ..."
        if line.startswith("#") and "| **Tools**:" in line:
            parts = line[1:].strip().split("| **Tools**:")
            project_name = parts[0].strip()
            tools_line = parts[1].strip()
            continue

        # Skip omitted skills
        if "(Skill omitted" in line:
            continue

        # Match actual bullet lines
        match = re.match(r"\*?([A-Za-z &\-()]+)\*?:\s*(.+)", line)
        if match:
            skill = match.group(1).strip()
            desc = match.group(2).strip().rstrip(".")
            bullets.append(f"- *{skill}*: {desc}.")

    return {
        "project_name": project_name,
        "tools": tools_line,
        "bullets": bullets
    }

from IPython.display import Markdown

output_lines = []

for project_name, raw_output in raw_results.items():
    parsed = extract_project_bullets_from_combined_header(raw_output)

    # Use bold formatting for project name and italic for tools
    output_lines.append(f"\n### **{parsed['project_name']}** | *Tools*: {parsed['tools']}*\n")

    # Use unordered list for bullet points with indentation
    for bullet in parsed["bullets"]:
        output_lines.append(f" {bullet}")

    output_lines.append("\n---\n")  # Separator between projects

# Join the lines into one final string
formatted_project_output = "\n".join(output_lines)
# print(formatted_project_output)
# Display the formatted output as Markdown
display(Markdown(formatted_project_output))
````
